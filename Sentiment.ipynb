Sentiment Analysis

```python
nltk.download('vader_lexicon')
```

```python
#Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()
```

```python
#Apply the sentiment analyzer to each headline and calculate the sentiment polarity
sentiments = []
for headline in data['news']:
  if pd.isnull(headline): #Check for null values
    polarity = 0.0 #Assign a neutral polarity for null values
  else:
    headline_str = str(headline) #Convert headline to a string
    polarity = sia.polarity_scores(headline_str)['compound']
  sentiments.append(polarity)

data['polarity'] = sentiments #Add the sentiment polarity values to the DataFrame

print(data) #Display the DataFrame with sentiment polarity values
```

```python
data.dropna(subset=['news'], inplace=True)
```

```python
plt.hist(data['polarity'], bins=20, edgecolor='black')
plt.xlabel('polarity')
plt.ylabel('Frequency')
plt.title('Distribution of Sentiment Polarity')
plt.show()
```

```python
from nltk.corpus import stopwords
from wordcloud import WordCloud
```

```python
#Download the required resourses from nltk
nltk.download('stopwords')
```

```python
#Combine all the headlines into a single string
all_headlines = ''.join(data['news'])

#Create a set of stopward
stopwords = set(stopwords.words('english'))
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(all_headlines)
```

```python
#Plot the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
